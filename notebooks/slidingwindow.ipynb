```python
import dask.dataframe as dd
import pandas as pd

# Example data
pdf = pd.DataFrame({"x": range(10)})
ddf = dd.from_pandas(pdf, npartitions=3)

window = 3

def rolling_mean(s):
    # This function runs on a *Pandas* Series in each partition
    return s.rolling(window=window, min_periods=1).mean()

result = ddf.x.map_overlap(
    rolling_mean,
    before=window - 1,   # rows of overlap before each partition
    after=0,             # rows of overlap after (often 0 for trailing windows)
    meta=('x', 'float64')
)
```
